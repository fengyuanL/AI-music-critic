```python
DataDict = {}
pieces = []
midi_path = "/content/drive/MyDrive/ECE324/GMP"
midi_detail_path = "/content/drive/MyDrive/ECE324/full_music_pieces_youtube_similarity_pianosoloprob_split.csv"
error_count = 0

with open('/content/drive/MyDrive/ECE324/spotifysearch.txt', 'w') as f:
  for root, dirs, files in os.walk(midi_path):
    for count, filename in enumerate(files):
      KEY = filename[:filename.rfind(", ")]
      DataDict[KEY] = {}
      index1 = filename.find(", ")
      index2 = filename[index1+2:].find(", ")
      index3 = filename[index1+index2+4:].rfind(", ")
      index4 = filename.rfind(".mid")
      DataDict[KEY]["File_Name"] = filename
      DataDict[KEY]["Surname"] = filename[:index1]
      DataDict[KEY]["First_Name"] = filename[index1+2:index1+index2+2]
      DataDict[KEY]["Piece_Name"] = filename[index1+index2+4:index1+index2+index3+4]
      DataDict[KEY]["Similarity"] = filename[index1+index2+index3+6:index4]
      pieces.append(filename[index1+index2+4:index1+index2+index3+4])
      f.write((filename[:index1] + "\t" + filename[index1+2:index1+index2+2] + "\t" + filename[index1+index2+4:index1+index2+index3+4] + "\n"))
    
```
